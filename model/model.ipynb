```python
# This defines a function called 'cross' that takes two column names as input.
def cross(Col1, Col2):
    # This line creates a cross-tabulation (contingency table) of the two specified columns.
    res = pd.crosstab(df[Col1], df[Col2])
    # Calculating the percentage of defaulters
    # This line calculates the percentage of defaulters within each category of the first column and stores it in a new column 'Percentage'.
    res['Percentage'] = round((res[res.columns[1]]/(res[res.columns[0]] + res[res.columns[1]])) * 100,2)
    # This line prints the resulting cross-tabulation table.
    print(res)
    #Plotting a stack bar graph
    # This line creates a plotly figure with two bar traces, one for non-defaulters and one for defaulters, stacked on top of each other.
    fig = go.Figure(data=[
    go.Bar(name='Non-Defaulters', x=res.index.to_list(),y=res[0]),
    go.Bar(name='Defaulters', x=res.index.to_list(), y=res[1],text=(res['Percentage']),textposition = "outside")])
    # Change the bar mode to stack
    # This line updates the layout of the figure, setting the bar mode to 'stack', the title, and the x and y axis titles.
    fig.update_layout(barmode='stack',title_text = res.index.name + " variable v/s target" ,xaxis_title=res.index.name,yaxis_title="Number of count")
    # This line displays the plotly figure.
    fig.show()
```

```python
df[['SEX','MARRIAGE','EDUCATION']] = df[['SEX','MARRIAGE','EDUCATION']].astype('object')

#One Hot encoding
df = pd.get_dummies(df)
df.head()
```

```python

# This line imports the train_test_split function from scikit-learn, used for splitting data into training and testing sets.
from sklearn.model_selection import train_test_split

# This line defines the features (X) by dropping the 'default' (target variable) and 'ID' columns from the DataFrame.
X = df.drop(['default','ID'], axis=1)
# This line defines the target variable (y) as the 'default' column from the DataFrame.
y = df['default']

# This line splits the data into training and testing sets. 70% of the data is used for training (X_train, y_train) and 30% for testing (X_test, y_test).
# 'stratify=y' ensures that the proportion of classes in the target variable is maintained in both sets.
# 'random_state=42' ensures that the split is the same every time the code is run, for reproducibility.
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, stratify=y, random_state=42)
```

```python
# create the training df by remerging X_train and y_train
df_train = X_train.join(y_train)
```

```python
# Separate majority and minority classes
# This line creates a new DataFrame 'df_majority' containing rows where the 'default' column is 0 (non-defaulters).
df_majority = df_train[df_train.default == 0]
# This line creates a new DataFrame 'df_minority' containing rows where the 'default' column is 1 (defaulters).
df_minority = df_train[df_train.default == 1]

# Print the counts of the majority and minority classes
# This line prints the number of non-defaulters in the training data.
print(df_majority.default.count())
print("-----------") # Separator for readability
# This line prints the number of defaulters in the training data.
print(df_minority.default.count())
print("-----------") # Separator for readability
# This line prints the counts of each unique value (0 and 1) in the 'default' column of the training data.
print(df_train.default.value_counts())
```

```python
from sklearn.utils import resample

# Upsample minority class
# This line uses the resample function to create a new DataFrame by sampling from the minority class ('df_minority').
df_minority_upsampled = resample(df_minority,
                                 replace=True,     # sample with replacement - This allows the same instance to be selected multiple times.
                                 n_samples=16355,   # to match majority class - The number of samples to generate, set to match the count of the majority class.
                                 random_state=42)   # reproducible results - Sets a seed for reproducibility of the sampling process.
# Combine majority class with upsampled minority class
# This line concatenates (joins) the original majority class DataFrame ('df_majority') with the newly upsampled minority class DataFrame ('df_minority_upsampled').
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
# Display new class counts
# This line prints the counts of each class in the combined, upsampled DataFrame to show the new balanced distribution.
df_upsampled.default.value_counts()
```

```python
# Downsample majority class
# This line uses the resample function to create a new DataFrame by sampling from the majority class ('df_majority').
df_majority_downsampled = resample(df_majority,
                                 replace=False,    # sample without replacement - Ensures each instance from the majority class is used at most once. [2, 3]
                                 n_samples=4645,     # to match minority class - The number of samples to select from the majority class, set to match the count of the minority class. [2]
                                 random_state=587) # reproducible results - Sets a seed for reproducibility of the sampling process.
# Combine minority class with downsampled majority class
# This line concatenates (joins) the newly downsampled majority class DataFrame ('df_majority_downsampled') with the original minority class DataFrame ('df_minority').
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
# Display new class counts
# This line prints the counts of each class in the combined, downsampled DataFrame to show the new balanced distribution.
df_downsampled.default.value_counts()
```

```python

from imblearn.over_sampling import SMOTE

# This line initializes the SMOTE object. SMOTE stands for Synthetic Minority Over-sampling Technique.
# It is a popular technique for handling imbalanced datasets by creating synthetic samples of the minority class.
# 'random_state' is set for reproducibility.
sm = SMOTE(random_state=42)
# This line applies SMOTE to the training data (X_train and y_train).
# 'fit_resample' learns the patterns in the minority class and generates synthetic samples.
# It returns two new arrays: X_SMOTE (features with synthetic samples) and y_SMOTE (target variable with corresponding synthetic labels).
X_SMOTE, y_SMOTE = sm.fit_resample(X_train, y_train)
# This line prints the total number of samples in the target variable after applying SMOTE, which will be increased due to the synthetic samples.
print(len(y_SMOTE))
# This line prints the sum of the target variable after applying SMOTE.
# Since the minority class is represented by 1, the sum will give the total count of the minority class after upsampling.
print(y_SMOTE.sum())
```

```python
# Define a function to evaluate machine learning classification models.
# It takes a model object and the training and testing data.
def model_eval(algo, Xtrain,ytrain,Xtest,ytest):

    # Import necessary functions from scikit-learn's metrics module
    # These are used to calculate various performance measures and create visualizations.
    from sklearn.metrics import confusion_matrix,accuracy_score,roc_auc_score,roc_curve,classification_report
    # Also need KFold for cross-validation
    from sklearn.model_selection import KFold

    # --- Evaluate Performance on Training Data ---
    # This step shows how well the model learned the patterns in the data it was trained on.

    # Train the provided machine learning algorithm using the training data (features Xtrain and target ytrain).
    algo.fit(Xtrain,ytrain)

    # Make predictions on the training data itself.
    # y_pred will contain the predicted class labels (e.g., 0 or 1) for the training samples.
    y_pred = algo.predict(Xtrain)

    # Get the predicted probabilities for the training data.
    # algo.predict_proba returns probabilities for each class. [:,1] selects the probabilities for the positive class (class 1).
    y_train_prob = algo.predict_proba(Xtrain)[:,1]

    # Print the overall accuracy on the training data.
    # Accuracy is the proportion of correctly predicted instances.
    print('Overall Train Accuracy',accuracy_score(ytrain,y_pred))

    # Print the AUC (Area Under the ROC Curve) score on the training data.
    # AUC measures the model's ability to distinguish between the positive and negative classes.
    print('Train AUC Score',roc_auc_score(ytrain,y_train_prob))

    # --- Evaluate Performance on Testing Data ---
    # This is the most important step. It shows how well the model performs on data it has never seen before,
    # which is a better indicator of real-world performance.

    # Make predictions on the testing data.
    # y_test_pred will contain the predicted class labels for the test samples.
    y_test_pred = algo.predict(Xtest)

    # Get the predicted probabilities for the testing data.
    # [:,1] selects the probabilities for the positive class (class 1).
    y_test_prob = algo.predict_proba(Xtest)[:,1]

    # Print the predicted probabilities for the test data, formatted as percentages.
    # This gives an idea of the model's confidence in its predictions.
    print("Test Probabilities (%):", np.round(y_test_prob * 100, 2))

    # Print the overall accuracy on the testing data.
    print('Overall Test Accuracy',accuracy_score(ytest,y_test_pred))

    # Print the AUC score on the testing data.
    print('Test AUC Score',roc_auc_score(ytest,y_test_prob))

    # Print a detailed classification report for the testing data.
    # This report includes Precision, Recall, F1-score, and Support for each class.
    print('Classification Report of Test\n',  classification_report(ytest, y_test_pred))

    # --- Perform K-Fold Cross-Validation on Training Data ---
    # K-Fold provides a more robust estimate of the model's performance than a single train/test split.
    # It helps in understanding the model's consistency and variability.

    # Initialize KFold cross-validation.
    # n_splits=5 means the training data will be split into 5 parts (folds).
    # shuffle=True shuffles the data before splitting.
    # random_state=42 ensures the shuffling is the same each time for reproducibility.
    kf = KFold(n_splits = 5,shuffle = True,random_state = 42)

    # Create an empty list to store the AUC score from each fold.
    score=[]

    # Loop through each fold generated by KFold.
    # kf.split() provides indices to split the data for training and testing within the cross-validation loop.
    for train_idx,test_idx in kf.split(Xtrain,ytrain):

        # Select the data for the current fold's training set using the provided indices.
        xtrain_k,xtest_k = Xtrain.iloc[train_idx,:],Xtrain.iloc[test_idx,:]
        # Select the corresponding target variable data for the current fold's training set.
        ytrain_k,ytest_k = ytrain.iloc[train_idx],ytrain.iloc[test_idx]

        # Train the model on the training data of the current fold.
        algo.fit(xtrain_k,ytrain_k)

        # Make predictions on the testing (validation) data of the current fold.
        y_pred_k = algo.predict(xtest_k)

        # Calculate the AUC score for the current fold's predictions.
        roc = roc_auc_score(ytest_k,y_pred_k)

        # Add the AUC score of the current fold to the list.
        score.append(roc)

    # Print the average AUC score across all K-Fold splits.
    # Also print the variance of the scores, which indicates how much the performance varies between folds.
    print('K-Fold scores: %0.03f (+/- %0.5f)' % (np.mean(score),np.var(score,ddof=1)))

    # --- Visualize Performance Metrics ---
    # Plotting the confusion matrix and ROC curve provides a visual understanding of model performance.

    # Create a figure and a set of subplots (1 row, 2 columns) for the plots.
    # figsize sets the size of the figure.
    f,ax =  plt.subplots(1,2,figsize=(14,6))

    # Calculate the confusion matrix for the test data.
    # This table summarizes the counts of True Positives, True Negatives, False Positives, and False Negatives.
    ConfMatrix = confusion_matrix(ytest,y_test_pred)

    # Create a heatmap visualization of the confusion matrix on the first subplot (ax[0]).
    # annot=True displays the actual numbers in each cell.
    # cmap sets the color scheme.
    # fmt="d" ensures the numbers are displayed as integers.
    # xticklabels and yticklabels set the labels for the axes.
    # linewidths adds lines between the cells.
    sns.heatmap(ConfMatrix,annot=True, cmap='YlGnBu', fmt="d",
            xticklabels = ['Non-default', 'Default'],
            yticklabels = ['Non-default', 'Default'],linewidths=.5,ax = ax[0])

    # Set the labels and title for the confusion matrix subplot.
    ax[0].set_ylabel('True label') # Actual class
    ax[0].set_xlabel('Predicted label') # Model's prediction
    ax[0].set_title('Confusion Matrix')

    # Calculate the data needed to plot the ROC curve.
    # fpr = False Positive Rate, tpr = True Positive Rate, thresholds are the probability thresholds used.
    # 'global' makes these variables accessible outside the function if needed (though generally avoided).
    global fpr,tpr,thresholds
    fpr,tpr,thresholds = roc_curve(ytest,y_test_prob)

    # Plot the ROC curve on the second subplot (ax[1]).
    # This plots TPR vs FPR at different thresholds. A curve closer to the top-left is better.
    ax[1].plot(fpr,tpr,color = 'r')

    # Plot the diagonal line representing a random classifier for comparison.
    ax[1].plot(fpr,fpr,color = 'green')

    # Set the labels and title for the ROC curve subplot.
    ax[1].set_ylabel('TPR') # True Positive Rate (Sensitivity)
    ax[1].set_xlabel('FPR') # False Positive Rate (1 - Specificity)
    ax[1].set_title('ROC Curve')

    # Display the plots.
    plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV, KFold

# Create lists to hold the training data and their corresponding names for different sampling strategies.
# These datasets were prepared in previous steps (original, oversampled, downsampled, SMOTE).
xtrain_data = [X_train,df_upsampled.drop('default',axis = 1),df_downsampled.drop('default',axis = 1),X_SMOTE]
ytrain_data = [y_train,df_upsampled['default'],df_downsampled['default'],y_SMOTE]
name = ['Normal Sampling' , 'Over Sampling' , 'Under Sampling' , 'SMOTE']

# Loop through each pair of training features (i) and target variables (j),
# along with the corresponding data name (k).
for i,j,k in zip(xtrain_data,ytrain_data,name):
    # Declare best_log as a global variable. Note: The way best_log is used here
    # might not be the most effective if the goal is to collect all best parameters;
    # it gets re-initialized in each loop iteration.
    global best_log
    # Print the name of the current data sampling strategy being used.
    print('Data is ',k)
    # Re-initialize the best_log list for each dataset. This means only the best
    # parameters from the last dataset processed will be stored in best_log
    # after the loop finishes.
    best_log = []
    # Setup the hyperparameter grid for Logistic Regression.
    # We are tuning the 'C' parameter, which is the inverse of the regularization strength.
    # A smaller 'C' means stronger regularization.
    # np.logspace(-5, 8, 15) creates 15 logarithmically spaced values for C between 10^-5 and 10^8.
    param_grid = {'C': np.logspace(-5, 8, 15)}
    # Instantiate a Logistic Regression classifier with default settings.
    logreg = LogisticRegression(max_iter=1000) # Added max_iter for convergence
    # Instantiate the RandomizedSearchCV object.
    # This object performs a random search over the hyperparameter grid.
    logreg_cv = RandomizedSearchCV(
        estimator=logreg,           # The model we want to tune
        param_distributions=param_grid, # The hyperparameters to search over
        scoring='roc_auc',          # The evaluation metric (Area Under ROC Curve)
        cv=5,                       # Number of cross-validation folds (5-fold CV)
        random_state=0              # Seed for reproducibility of the random search
    )
    # Fit RandomizedSearchCV to the current training data.
    # This trains the model multiple times with different hyperparameter combinations
    # using cross-validation to find the best combination.
    logreg_cv.fit(i,j)
    # Append the best parameters found for the current dataset to the best_log list.
    # As mentioned, this list is reset in each iteration, so it won't contain
    # the best parameters for all datasets at the end.
    best_log.append(logreg_cv.best_params_)
    # Print the best hyperparameters found by RandomizedSearchCV for the current dataset.
    print("Tuned Logistic Regression Parameters: {}".format(logreg_cv.best_params_))
    # Print a separator line for clarity.
    print("_" * 100)
```

```python
# Create a list containing the feature sets for each training dataset version.
# This includes the original training data, oversampled data (excluding the target),
# downsampled data (excluding the target), and SMOTE-generated data.
xtrain_data = [X_train,df_upsampled.drop('default',axis = 1),df_downsampled.drop('default',axis = 1),X_SMOTE]

# Create a list containing the corresponding target variable sets for each training dataset version.
ytrain_data = [y_train,df_upsampled['default'],df_downsampled['default'],y_SMOTE]

# Create a list of names for each sampling strategy, used for clear output.
name = ['Normal Sampling' , 'Over Sampling' , 'Under Sampling' , 'SMOTE']

# Create a list of indices to access the correct set of best hyperparameters
# from the 'param_log' list for each sampling strategy.
index = [0,1,2,3]

# Loop through each combination of training features, training target, sampling strategy name, and hyperparameter index.
for i,j,k,l in zip(xtrain_data,ytrain_data,name,index):
    # Print a message indicating which sampling strategy and corresponding hyperparameters are being used.
    print('Data is ',k,' And with hyper parameter ',param_log[l])

    # Evaluate a Logistic Regression model:
    # - The model is initialized with the best hyperparameters specific to the current sampling strategy (param_log[l]).
    # - The 'random_state' is set for reproducibility.
    # - The model is trained on the current training data (i, j).
    # - The model's performance is evaluated on the standard, unseen test data (X_test, y_test).
    model_eval(LogisticRegression(**param_log[l],random_state= 42), i,j,X_test,y_test)

    # Print a separator line for visual clarity between evaluations of different sampling strategies.
    print("_" * 60)
```
